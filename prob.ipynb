```python
import operator
from scipy.special import comb

def cal_num(score_num): 
    a = b = c = range(0, 51) 
    result = 0 
    for x in a: 
        for y in b: 
            for z in c: 
                if x + y + z == score_num: 
                    result += 1 
    return result

all = 2**100*51**3
result = {}

for score_num in range(0, 126):
    num = 0
    for score_num1 in range(0, 101):
        num_1 = comb(100, score_num1)
        if score_num1 <= score_num:
            num_2 = cal_num(score_num - score_num1)
            num += num_1 * num_2
    if score_num == 125:
        print 'Total number of scenarios %d' % all
        print 'Total number of tied-scenarios %d' % num
    key_name = 'Scenario ' + str(score_num) + ':' + str(250-score_num)
    result[key_name] = float(num)/all
        
result_sort = sorted(result.items(), key = operator.itemgetter(1), reverse = True)

for each in result_sort:
    print each[0], "\t", each[1]
```
